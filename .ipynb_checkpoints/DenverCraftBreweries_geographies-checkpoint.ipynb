## Geographic Boundaries
### For geospatial boundaries of our chosen area, the Denver Metro area, we turn to the US Census Bureau who has created various shapefiles for geographies throughout the US

https://www.census.gov/geographies/mapping-files/time-series/geo/tiger-geodatabase-file.html
* TIGER/Line Geodatabases are spatial extracts from the Census Bureau’s **Master Address File**/**Topologically Integrated Geographic Encoding and Referencing** **(MAF/TIGER)** System for use with geographic information systems (GIS) software. The geodatabases contain national coverage (for geographic boundaries or features) or state coverage (boundaries within state).
* These files do not include demographic data, but they contain geographic entity codes that can be linked to the Census Bureau’s demographic data, available on **data.census.gov**.
* This includes MSAs detailed below

### 2023 US Census Bureau Metropolitan-Micropolital Statistical Areas (MSA)
* https://www.census.gov/geographies/reference-files/time-series/demo/metro-micro/delineation-files.html
* We use the Census Bureau definition for the greater Denver Metro area, defined as the **Denver-Lakewood-Aurora** metropolitan area.
    - we will supplement this with cities, towns, etc. immediately adjacent to the MSA geospatial boundary, but not including Boulder
    - we are mainly interested in the adjacent area Northwest of the MSA where there is a sizeable population
* MSAs are more densely populated geographic areas found throughout the US
* MSAs are comprised of state counties which fully or partially encompass a statistical area
    - this means we may capture places outside of a metro area which are still w/in a large metro county where geographic pruning may be needed, e.g. eastern Arapahoe County
* US Census Bureau **delineations** breakdown an MSA
    - "A metropolitan or micropolitan statistical area's geographic composition, or list of geographic components at a particular point in time, is referred to as its **"delineation."** Metropolitan and micropolitan statistical areas are delineated by the U.S. Office of Management and Budget (OMB) and are the result of the application of published standards to Census Bureau data. The standards for delineating the areas are reviewed and revised once every ten years."
    - https://www.census.gov/geographies/reference-files/time-series/demo/metro-micro/delineation-files.html

```python
def find_folder_file(desired_folder, file_ext=None, pwd=False):
    """Given the main project directory recursively find the desired folder name, 
    and if provided, find files matching the file extension from within the desired folder; 
    there is commonly only 1 file of a given extension such as a shapefile '.shp' in a dataset due to 
    how the US Census Bureau structures data, but not always.
    NOTE: files are collectively found by os.walk() from the root directory
    and NOT directory-by-directory, unfortunately.
    params:
    desired_folder{str} - name of the desired folder
    file_ext{str} - desired file extension
    pwd{bool} - if True will print the present working directory
    returns:
    os.listdir(os.getcwd()){list} - returns a list of matching file extensions for the desired directory
    """
    # use os.walk to recurisvely scan directories inside of top-level directory provided
    for root, dirs, files in os.walk('/home/jcorley60/Documents/project_portfolio/DenverCraftBreweries', 
                                     topdown=True):
        for dirname in dirs:
            if dirname == desired_folder:
                os.chdir(os.path.join(root, dirname))
    if pwd:
        print(os.getcwd())
    
    # if file extension provided then proceed to list contents of desired directory
    # use os.getcwd() to avoid n**3 above (n**2 + n instead where n is contents of single directory)   
    files = []
    if file_ext is not None:
        files += [file for file in os.listdir(os.getcwd()) if file.endswith(file_ext)]
        return files    
    
    elif file_ext is None:
        return os.listdir(os.getcwd())
        
                
find_folder_file('MSA', 'xlsx', True)
```

### Get Denver Metro MSA
* Delineation of central vs an outlying county is important for our MSA geography
* we will use only 'Central' counties found in the Denver MSA as for example the 'Outlying' county of Gilpin is counted within the Denver-Lakewood-Aurora MSA, however, Gilpin County is far removed from the approximate MSA area in that it is geographically separated by the Front Range mountains.

```python
def get_msa(natl_msa_df, msa_area):
    """Finds locales comprising the given Metropolitan/Micropolitan statistical (MSA) area
    params:
    natl_msa{Pandas DF} - DF of MSA's of the the country
    msa_area{str} - A locale name for the desired MSA
    returns:
    metro_counties{Pandas DF} - counties for the desired MSA
    df_denver{Pandas DF} - Denver MSA counties
    """
    df_denver = natl_msa_df.loc[natl_msa_df['CBSA Title'].str.contains(msa_area, na=False)]

    # only use 'Central' counties for Metropolitan/Micropolitan Statistical Area
    # For example Gilpin County is considered an 'Outlying County'
    # but encompasses Central City & Black Hawk is isolated due to mountains & irrelevant
    df_denver = df_denver[df_denver['Central/Outlying County'] == 'Central']

    metro_counties = df_denver['County/County Equivalent'].to_numpy()
    print(f"{len(metro_counties)} counties reported w/in {msa_area} MSA:\n{[i for i in metro_counties]}")

    return metro_counties, df_denver


metro_counties, df_denver = get_msa(df_msa, 'Denver')
df_denver
```

#### Find only MSA Denver-Lakewood-Aurora Metro Counties
* This will enable us to determine Census Bureau `Places` w/in our county boundaries
* The liberty of using 2022 county geospatial boundaries has been taken as no corresponding 2023 geospatial data could be immediately located

```python
def get_msa_counties(state, msa_counties):
    """Given US counties and a list of desired MSA counties, finds geometry for counties of the MSA from
    US Census Bureau US counties file.
    params:
    state{str} - State(s) that the MSA is found within
    msa_counties{str} - list of desired MSA counties
    returns:
    gdf_metro_counties{GeoPandas GeoDataFrame} - GeoDataFrame (gdf) of metro counties
    """
    
    # gdf meaning GeoDataFrame - contains geospatial coordinate pairs (lat, lon)
    gdf_counties = gpd.read_file(find_folder_file('cb_2022_us_county_500k', 'shp', True)[0]) # US counties
    
    # there are frequently same-name counties in other states, for example Jefferson County, KS vs CO
    gdf_counties = gdf_counties[gdf_counties['STATE_NAME'] == state]    # Colorado state counties
    
    # Find MSA counties from US counties
    gdf_metro_counties = gdf_counties[gdf_counties['NAMELSAD'].isin(msa_counties)].sort_values(by='NAMELSAD')
    print(gdf_metro_counties.shape)
    
    return gdf_metro_counties


gdf_metro_counties = get_msa_counties('Colorado', metro_counties)
gdf_metro_counties
```

```python
def plot_geography(
    geoDataFrame_geom, 
    chart_title, 
    x=8, 
    y=6, 
    grid_alpha=None, 
    cmap=None, 
    color=None, 
    ha='right', 
    rotation=45,
    mult_plot=False,
    ax=None
):
    """Function to plot geographic locations from latitude and longitude coordinates.
    params:
    geodataframe_geom{GeoPandas GeoDF} - required
    chart_title{str}              - required
    x{int}                        - desired length of plot
    y{int}                        - desired heighth of plot
    grid_alpha{float}             - controls if background grid is present and its intensity w/in graph
    cmap{str}                     - color map, or mulitple colors in a package e.g. 'viridis' or 'magma'
    color{str}                    - instead of a color map, a single color can be used for a plot
    ha{'right', 'center', 'left'} - horizontal alignment of x-axis labels
    rotation{int}                 - angle of x-axis labels
    mult_plot{True/False}         - multiple plots
    ax{axis-object}               - Matplotlib axis object used for construction of plot
    
    returns:
    none - Matplotlib plot
    """
    if mult_plot is False:                   # mult_plot: True/False, multiple plots in same output
        fig = plt.figure(figsize=(x, y))    # Adjust the figure size values as needed
        ax = fig.gca()                      # create an axis for plotting - get current axis (GCA)  
        
    geoDataFrame_geom.plot(ax=ax, cmap=cmap, color=color);
    plt.title(chart_title);
    plt.xlabel('Longitude');
    plt.ylabel('Latitude');
    if grid_alpha is not None:
        plt.grid(which='major', axis='both', alpha=grid_alpha); 
    plt.xticks(ha=ha, rotation=rotation)
```

```python
plot_geography(
    gdf_metro_counties, 
    'Denver-Aurora-Lakewood MSA Counties', 
    x=7,
    y=5,
    ha='center',
    rotation=0
)

plt.show()
```

### Consolidate geometries
Keep only the original outer boundary for all 6-counties in order to enable selection of Census Bureau `Places` that fall within the consolidated county/MSA area

```python
def consolidate_geometries(gdf):
    """Consolidates any inner geometric boundaries leaving the outer-most boundary.
    params: gdf{gdf to be consolidated} - dissolves inner geometries but preserves outermost boundary
    returns: gdf_merge{gdf} - gdf of merged geometries
    """
    # when merging geometries a single row from the 1st values is kept, apart from the geometry column
    temp = gdf[['geometry']]   
    gdf_merge = temp.dissolve()    # dissolve inner geometric/geospatial boundaries
    return gdf_merge


gdf_county_merge = consolidate_geometries(gdf_metro_counties)
gdf_county_merge
```

```python
plot_geography(
    gdf_county_merge['geometry'], 
    'Denver-Aurora-Lakewood MSA\nConsolidated County Boundary', 
    x=7,
    y=6,
    grid_alpha=0.2,
    ha='center',
    rotation=0
)

plt.show()
```

## Census `Places` 
* https://www.census.gov/geographies/mapping-files/time-series/geo/cartographic-boundary.2022.html#list-tab-1883739534
* This identifies places within the specified geographic region including local cities, towns, etc.
* NOTE: between 2022 & 2023 the 6-MSA counties did not change their geographic boundaries as determined by Googling county changes
* Places follow ANSI derived FIP codes which are used as the unique identifier for each geospatial area

```python
def get_colorado_places():
    """Function to read in places found w/in Colorado (Census Bureau State id: 08).
    params: none
    returns: gdf_places{gdf} - GeoDataFrame of Census Places found w/in Colorado
    """
    gdf_places = gpd.read_file(find_folder_file('cb_2022_08_place_500k', 'shp', True)[0])

    # "LSAD" stands for "Legal/Statistical Area Description."
    # It is a code used to classify and describe various geographic and administrative entities in the United States
    gdf_places = gdf_places[['AFFGEOID', 'NAME', 'NAMELSAD', 'ALAND', 'AWATER', 'geometry']]
    
    return gdf_places
    
    
gdf_places = get_colorado_places()
gdf_places
```

### Locate Census Bureau `Places` within the Denver `MSA`
* Geopandas method **.sjoin** parameter *predicate* will determine the geospatial operation to be performed; we use intersect
    - **intersects** will find all geometries which completely reside w/in the reference geometry **or** are adjacent to it
    - *within* will find geometries which completely reside w/in the reference geometry, only
* We would like to capture sizeable population densities within the 6-county MSA, or immediately adjacent to

```python
def get_msa_places(places, county_boundary, predicate='intersects'):
    """Finds Census Bureau Places w/in the MSA consolidated county area
    params:
    places{gdf} - GeoDataFrame of desired MSA places
    county_boundary{gdf} - GeoDataFrame of outer county boundary
    predicate{'intersects' or 'within'} - intersects means Census Places must completely 
    reside w/in county_boundary; within means the Census Places can reside w/in or touch the county_boundary
    returns:
    gdf_msa_places{gdf} - GeoDataFrame of Census Places w/in county_boundary
    """
    gdf_msa_places = gpd.sjoin(places, county_boundary, predicate=predicate)
    gdf_msa_places.drop(columns=['index_right'], inplace=True)
    gdf_msa_places.reset_index(inplace=True, drop=True)
    gdf_msa_places.rename({'AFFGEOID': 'GEO_ID'}, axis=1, inplace=True)
    
    return gdf_msa_places
    

gdf_msa_places = get_msa_places(gdf_places, gdf_county_merge)
gdf_msa_places
```

```python
fig = plt.figure(figsize=(8, 6))  # Adjust the figure size values as needed
ax = fig.gca()                     # create an axis for plotting - get current axis (GCA)

plot_geography(
    gdf_metro_counties, 
    'Denver-Lakewood-Aurora MSA:\nCounties & Their Associated Places', 
    grid_alpha=0.3,
    ha='center',
    rotation=0,
    mult_plot=True,
    ax=ax,
    color='darkblue'
)

plot_geography(
    gdf_msa_places, 
    'Denver-Lakewood-Aurora MSA:\nCounties & Their Associated Places', 
    grid_alpha=0.3,
    ha='center',
    rotation=0,
    mult_plot=True,
    ax=ax,
    cmap='coolwarm'
)

plt.show()
```

### AREALM
* https://www2.census.gov/geo/tiger/TIGER_RD18/STATE/08_COLORADO/08/
* AREALM - Area Landmarks
* hospitals, colleges, nursing homes, municpal centers, public parks, correctional facilities, cemeteries, golf courses, Walmart retail locations, Sheriff's offices, etc.

#### Important Areas & Point Landmarks & Their Abbreviations
* Analysis document:
    - to determine which keywords were needed, an analysis of area and place landmarks was carried out
    - see `analysis_areas_places.ods` in /DenverCraftBreweries/data/US Census Bureau
* Search strings:
    - Colleges (Colg), or Universities (Univ, University, School)
    - Hospitals (Hosp, Hospital) or Medical Centers

```python
def get_census_areas_within_msa(gdf_msa_places):
    """Function to find Census Bureau area (polygon) landmarks which fall w/in the MSA.
    Returns:
    merged geodataframe where the left half of the GeoDataFrame represents the area landmark, and 
    the right captures the associated Census Place info where the landmark is located.
    params: gdf_msa_places{gdf} - GeoDataFrame of desired Census Places
    returns: gdf_msa_realm{gdf} - GeoDataFrame containing area landmark geometries
    """
    
    gdf_realm = gpd.read_file(find_folder_file('tl_rd22_08_arealm', 'shp', True)[0])

    gdf_msa_realm = gpd.sjoin(gdf_realm, gdf_msa_places, predicate='intersects')  # lsuffix='', rsuffix='_'
    gdf_msa_realm.drop(columns=['index_right'], axis=1, inplace=True)

    # grab area landmarks which only have names
    gdf_msa_realm = gdf_msa_realm[gdf_msa_realm['FULLNAME'].notna()]

    # Select locations (locales which aid in supporting a craft brewery):
    # colleges, airports, and medical centers
    search_str = 'colg|college|univ|university|mines|arprt|hospital|hosp|medical center|medical ctr|med center|plz'
    gdf_msa_realm = gdf_msa_realm[gdf_msa_realm['FULLNAME'].str.contains(search_str, case=False)]

    # drop row for false positive match
    # drop 'The Childrens Hospital Association' it's just the assoc, whereas the others no longer exist
    drop_list = [
    'Belleview Christian Colg',
    'Centura Health-St Anthony Central Hosp', 
    'Clayton Colg',
    'Hospice of Saint John Ltc',
    'Johnson & Wales University Denver', 
    'Mercy Med Center Hosp',
    'The Childrens Hospital Association'
    ]

    gdf_msa_realm = gdf_msa_realm[~gdf_msa_realm['FULLNAME'].isin(drop_list)]

    # drop columns
    gdf_msa_realm.drop(columns=[
        'STATEFP', 
        'ANSICODE', 
        'AREAID', 
        'MTFCC', 
        'PARTFLG',
        'NAME', 
        'ALAND_right', 
        'AWATER_right'
    ], axis=1, inplace=True)

    # drop duplicates -- there will be several places w/ multiple assoc. areas, however, we just need 1
    gdf_msa_realm.drop_duplicates(subset='FULLNAME', keep='first', inplace=True)
    gdf_msa_realm.drop(columns=['num_major_roads'], axis=1, inplace=True)
    
    gdf_msa_realm.sort_values(by='FULLNAME', inplace=True)
    gdf_msa_realm.reset_index(inplace=True, drop=True)
    
    # Write to file for backup
    os.chdir('/home/jcorley60/Documents/project_portfolio/DenverCraftBreweries/\
data/US Census Bureau/tl_rd22_08_arealm')
    gdf_msa_realm.to_pickle('gdf_msa_realm.pkl')

    return gdf_msa_realm


gdf_msa_realm = get_census_areas_within_msa(gdf_msa_places)
gdf_msa_realm.head()
```

### Create columns denoting the number of colleges or hospitals found w/in a Census Bureau Place

```python
# There are only 2 airports listed in this dataset--places [below] has these and additional heliports

def column_filler_has_college(row):
    """Helper function to determine colleges"""
    return 1 if any(keyword in row['FULLNAME'].lower() for keyword in [
        "colg", "college", "univ", "university", "mines"]) else 0

def column_filler_has_hospital(row):
    """Helper function to determine medical centers
    IMPORTANT NOTE: not included are doctors offices nor urgent care facilities"""
    return 1 if any(keyword in row['FULLNAME'].lower() for keyword in [
        "hospital", "hosp", "medical center", "medical ctr", "med center", "plz"]) else 0

      
gdf_msa_realm['num_colleges'] = gdf_msa_realm.apply(column_filler_has_college, axis=1)
gdf_msa_realm['num_medical_facilities'] = gdf_msa_realm.apply(column_filler_has_hospital, axis=1)
```

#### Perform aggregation via groupby

```python
gdf_msa_realm_groupby = gdf_msa_realm.groupby('GEO_ID').agg({
    'NAMELSAD': 'first',         # keep the first NAMELSAD (AKA name) for each city associated w/ GEO_ID
    'num_colleges': 'sum',           # Sum 'colleges' values for each city
    'num_medical_facilities': 'sum', # Sum the 'medical_facilities' (hospitals, medical ctrs) values for each city
}).reset_index()

gdf_msa_realm_groupby
```

### Merge `gdf_msa_realm_groupby` w/ `gdf_msa_places` to obtain college & medical facility counts per Census Place

```python
gdf_msa_places = pd.merge(gdf_msa_places, gdf_msa_realm_groupby, how='left')
gdf_msa_places.fillna(0, inplace=True)
gdf_msa_places[['num_colleges', 'num_medical_facilities']] = gdf_msa_places[[
    'num_colleges', 'num_medical_facilities']].astype('int')

gdf_msa_places
```

### pointlm
* https://www2.census.gov/geo/tiger/TIGER_RD18/STATE/08_COLORADO/08/
* pointlm - Point Landmarks
* this captures similar info to Area Landmarks but instead of a polygon our geometry is exclusively comprised of points
#### Important Areas & Point Landmarks & Their Abbreviations
* Analysis document:
    - to determine which keywords were needed, an analysis of area and place landmarks was carried out
    - see `analysis_areas_places.ods` in /DenverCraftBreweries/data/US Census Bureau
* Search strings:
    - Airports (Arprt)

```python
def get_census_points_within_msa(gdf_msa_places):
    """Function to find Census Bureau point landmarks which fall w/in the given MSA.
    params: gdf_msa_places{gdf} - GeoDataFrame of desired Census Places w/ respective geometries.
    returns:
    merged geodataframe where the left half of the GeoDataFrame represents the point landmark, and 
    the right captures the associated Census Place info where the landmark is located.
    """
    
    gdf_pointlm = gpd.read_file(find_folder_file('tl_rd22_08_pointlm', 'shp', True)[0])

    # Drop rows which have an NA value for 'FULLNAME' column (we won't know what the row represents)
    gdf_pointlm = gdf_pointlm[gdf_pointlm['FULLNAME'].notna()]

    # Geospatial join to determine which points lie w/in MSA
    gdf_msa_pointlm = gpd.sjoin(gdf_pointlm, gdf_msa_places, predicate='intersects')  # lsuffix='', rsuffix='_'

    # Limit results to desired occupation-related points
    search_str = "colg|arprt|afb|hosp|med center|medical"
    gdf_msa_pointlm = gdf_msa_pointlm[gdf_msa_pointlm['FULLNAME'].str.contains(search_str, case=False)]

    gdf_msa_pointlm.drop(columns=[
        'POINTID', 
        'STATEFP', 
        'ANSICODE', 
        'MTFCC', 
        'index_right', 
        'NAME'
    ], axis=1, inplace=True)
    
    # these hospitals appear to be permanently closed -- drop them
    hospitals_to_exclude = ['St Anthony Hosp', 'Beth Isreal Hosp', 'Porter Memorial Hosp']
    gdf_msa_pointlm = gdf_msa_pointlm[~gdf_msa_pointlm['FULLNAME'].isin(hospitals_to_exclude)]
    gdf_msa_pointlm.drop(columns=[
        'num_major_roads', 
        'num_colleges', 
        'num_medical_facilities',
        'ALAND',
        'AWATER'
    ], axis=1, inplace=True)

    gdf_msa_pointlm.sort_values(by=['FULLNAME'], inplace=True)
    gdf_msa_pointlm.reset_index(inplace=True, drop=True)
    
    # Write to file for backup
    os.chdir('/home/jcorley60/Documents/project_portfolio/DenverCraftBreweries/\
data/US Census Bureau/tl_rd22_08_pointlm')
    gdf_msa_pointlm.to_pickle('gdf_msa_pointlm.pkl')

    return gdf_msa_pointlm


gdf_msa_pointlm = get_census_points_within_msa(gdf_msa_places)
gdf_msa_pointlm
```

### Create new columns denoting if Census Place has airport
* data for hospitals/medical centers above appears very close in accuracy to this set and is not presented again here
* data for colleges in this dataset is lacking compared to set above and in reality. The dataset above for areas is used instead.

```python
def column_filler_has_airport(row):
    """We will include Buckley AFB under Aurora under `airports`
    as Buckley resides w/in the city limits of Aurora, CO."""
    
    return 1 if any(keyword in row['FULLNAME'].lower() for keyword in ["arprt", "afb"]) else 0


gdf_msa_pointlm['num_airports'] = gdf_msa_pointlm.apply(column_filler_has_airport, axis=1)
```

### Consolidate results by column `GEO_ID` to arrive at totals per `GEO_ID`

```python
gdf_msa_pointlm_groupby = gdf_msa_pointlm.groupby('GEO_ID').agg({
    'NAMELSAD': 'first',          # keep the first NAMELSAD (AKA name) for each city associated w/ GEO_ID
    'num_airports': 'sum'             # Sum the 'airport_heliports' (# of airports or heliports) values for each city
}).reset_index()

gdf_msa_pointlm_groupby
```

### Merge Point Landmark Groupby GeoDataFrame w/ `gdf_msa_places` to capture landmarks per Census Place

```python
gdf_msa_places = pd.merge(gdf_msa_places, gdf_msa_pointlm_groupby, how='left')
gdf_msa_places.fillna(0, inplace=True)
gdf_msa_places['num_airports'] = gdf_msa_places['num_airports'].astype('int')
gdf_msa_places
```
