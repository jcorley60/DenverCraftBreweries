## Population Density for Each City

### Open Merged JSON File & Extract Data

```python
with open('METRO_BREWERIES.json', 'r') as file:
    venue_data = json.load(file)


name_list, lat_list, lon_list, category_list = [], [], [], []

for i in range(0, len(venue_data)):
    name = venue_data[i]['name']
    name_list.append(name)
    
    lat = venue_data[i]['location']['lat']
    lat_list.append(lat)
    
    lon = venue_data[i]['location']['lng']
    lon_list.append(lon)
    
    category = venue_data[i]['categories'][0]['name']
    category_list.append(category)
    
venue_df = pd.DataFrame(
    {
        'Name': name_list, 
        'Category': category_list,
        'Lat': lat_list, 
        'Lon': lon_list, 
    }
)

venue_df.shape
```

## Determining Which Neighborhoods Have the Most Breweries.

Figuring out if a point (lat/lon) lies within a 2-D geometry (collection of lats/lons) is known as Point-in-Polygon (PIP).  The Foursquare latitude/longitude coordinate pair need a Coordinate Reference System; reviewing their Developer documentation I did not find a mention of this, and have therefore assumed among the most common in the USA common WGS84 (EPSG 4326).

### Format Data

```python
# Point Cases
venue_df
venue_gdf = gp.GeoDataFrame(venue_df, geometry=gp.points_from_xy(x=venue_df.Lon, y=venue_df.Lat, crs="EPSG:4326"))

venue_gdf.drop(columns=['Category', 'Lat', 'Lon'], inplace=True)
print(venue_gdf.head())


## Polygon Areas
gdf_cleaned = gp.GeoDataFrame(gdf, columns=['NAME', 'geometry'])

print('\n', gdf_cleaned.head())

print('\nto_epsg()', venue_gdf.crs.to_epsg())
print('venue_gdf CRS: ', venue_gdf.crs)
print('gdf_cleaned CRS: ', gdf_cleaned.crs)

print("'venue_gdf.crs' = 'gdf_cleaned.crs'? :", venue_gdf.crs == gdf_cleaned.crs)
```

## Density-Based Clustering of Metro Area Breweries

#### Density-based cluster was chosen due to the nature of the dataset:
DBSCAN stands for Density-Based Spatial Clustering of Applications with Noise.  Density-based Clustering locates regions of high [cluster] density and separates them from other regions with low density. Density, in this context, is defined as the number of points within a specified radius and is a comparative measure.  DBSCAN is an unsupervised Machine Learning algorithm.  DBSCAN is unimpacted by noise.

### Preprocess data for use with Scikit-Learn DBSCAN

```python
df_venues_dbscan['NAME']

dbscan_df = venue_df
# dbscan_df.drop(columns=['geometry', 'Category'], inplace=True)
dbscan_df['NAME'] = df_venues_dbscan['NAME']
dbscan_df.shape
dbscan_df.head()
```

```python
df = pd.merge(dbscan_df, final_df, on=['NAME'], how='left')
df.drop(columns=['INTPTLAT', 'INTPTLON', 'NAME', 'Brewery Count', 'Population', 'ALAND', 'Name', 'Population Density', 'geometry', 'Category'], inplace=True)
print(df.shape)
print(df.head())

coords = df.to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN


df_clust = StandardScaler().fit_transform(np.radians(coords))
dbs = DBSCAN(eps=0.5, min_samples=2, metric='haversine').fit(df_clust) 
cluster_labels = dbs.labels_ 
df["Clus_Db"] = cluster_labels

num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)

print(f"""
Number of Real Clusters: {num_clusters}
Cluster Labels: \n{cluster_labels}
""")
```

```python
# Building the label to color mapping 

colors1 = ['r', 'g', 'b', 'c', 'y', 'm', 'w', 'k'] 
cvec1 = [colors1[label] for label in cluster_labels] 
  
r = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='r') 
g = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='g') 
b = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='b') 
c = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='c') 
y = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='y') 
m = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='m')
w = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='w')
k = plt.scatter(df['Lon'], df['Lat'], marker ='o', color='k') 
```
